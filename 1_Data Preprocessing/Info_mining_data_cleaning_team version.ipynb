# Data Preprocessing

```python
import pandas as pd
import re
```

```python
#df = pd.read_csv('NewsFakeCOVID-19.csv', encoding='ansi')
df = pd.read_csv('NewsRealCOVID-19.csv', encoding='ansi')

#df.fillna
df.tail(10)
```

```python
df2=df[['type','content']]
df2
```

```python
#filter out NaN/null
filtered_df2 = df2[df2['content'].notnull()] 
filtered_df2
```

```python
import nltk
from nltk import word_tokenize, Text
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer 
from nltk.stem import WordNetLemmatizer 

def remove_punctuation(line):
    return re.sub(r'[^\w\s]', '', line) 

#ps = PorterStemmer()     
lemmatizer = WordNetLemmatizer()
info_fake=[]
for a in filtered_df2['content']:
    word=remove_punctuation(a)
    word_ = word.replace('Ã','')
    new_word=word_.lower()
    #print(new_word)
    tokens = word_tokenize(new_word)
    all_stopwords=stopwords.words('english')
    tokens_no_stopw = [word for word in tokens if not word in all_stopwords]
    #print(tokens_no_stopw)
    stem_s = ' '.join(lemmatizer.lemmatize(w) for w in tokens_no_stopw)
    #print(stem_s)
    info_fake.append(stem_s)
   
content_cleaned = pd.DataFrame({'content_cleaned' : info_fake},
    
                    columns=['content_cleaned'])
```

```python
filtered_df2 = filtered_df2.assign(content_cleaned=content_cleaned.values)
filtered_df2
```

```python
fake_news=filtered_df2[['type','content_cleaned']]
fake_news
```

```python
#export in the form of csv
fake_news.to_csv('fake_news_cleaned.csv')
```
